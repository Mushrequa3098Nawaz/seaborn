```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data=sns.load_dataset('titanic')
```

```python
data.head()
```

```python
# Problem 1 : Plot a Heatmap on the correlation between multiple columns
sns.heatmap(data.corr())
#the heatmap provides a colored visual summary of the information given in the dataset representing it in a 2-dimensional form
```

```python
# Problem 2 : Plot the distribution of age column using a Boxplot on the basis of sex column
sns.catplot(x='sex', y='age', kind='box', data=data)
#a boxplot is useful when we want to compare data between two groups.
```

```python
sns.boxplot(x='sex', y='age', data=data)#gives the same result
```

```python
# Problem 3 : Plot the distribution of age using a ViolinPlot on the basis of both Pclass and Sex column
sns.catplot(x='sex', y='age', kind='violin', hue='pclass', data=data)
sns.catplot(x='pclass', y='age', kind='violin', hue='sex', data=data)
#Violin Plot is a method to visualize the distribution of numerical data of different variables.
```

```python
# Problem 4 : Plot the count of Survivors and Non-Survivors
sns.countplot(x="alive", data=data)
sns.catplot(x='alive', kind='count', data=data)#gives the same result
```

```python
# Problem 5: Plot the count of survivors as well as non-survivors on the basis of Pclass
sns.countplot(x="alive", hue='pclass',data=data)
```

```python
# Problem 6: Plot a scatterplot between age and fare column on the basis of both Pclass and Sex column
sns.relplot(x='age',y='fare',style='sex',hue='pclass',data=data)
#this scatterplot shows the different class of the passengers on the titanic and whether the passenger is a male or a female.
```

```python
sns.scatterplot(x='age',y='fare',style='sex',hue='pclass',data=data)#gives the same result
```
